# Accurate Toeplitz Matrix Calcurate

```julia
using PyCall
using PyPlot

scipy_ndimage = pyimport("scipy.ndimage")

function Fourier_permittivity(Nx, Ny, x_start, x_end, y_start, y_end, n, angle)
    
    A = ones(Float64, Nx, Ny) 

    A[y_start + 1:y_end, x_start + 1:x_end] .= n
    #A[x_start:x_end, y_start:y_end] .= n

    # Calculate the center of the rectangle
    center_x = div(x_start + x_end, 2)
    center_y = div(y_start + y_end, 2)

    # Rotate rectangle around its center using scipy's rotate function
    A_rotated = scipy_ndimage.rotate(A, angle, reshape=false, order=1, cval=1)

    # permittivity figure
    PyPlot.imshow(A_rotated, cmap="viridis", origin="lower")
    PyPlot.colorbar(label="permittivity")
    PyPlot.xlabel("x")
    PyPlot.ylabel("y")
    PyPlot.grid(true, linestyle="dotted", color="white")
    PyPlot.title("Permittivity distribution")
    PyPlot.show()

    return A_rotated
end
```

# Permittivity distribution

```julia
# parameters
Rex = 1200
Rey = 1200
y_start = 350
x_start = 200
y_end = 850
x_end = 1000
n = 1.6^2
angle = 0

F_1 = Fourier_permittivity(Rex, Rey, x_start, x_end, y_start, y_end, n, angle);
```

```julia
using FFTW

F_r = 1 ./ F_1;
```

### -2N ~ 2N まで$\frac{1}{\epsilon(x,y)}$の$x_p$点のフーリエ係数をyに関して計算し、Toeplitz行列に変換

### $\alpha(x_p)は$$x_p$におけるTpeplitz行列の逆行列を1行にしたもの

### $\alpha(x_p)$を一行に直して以下のように縦に並べる
$ [\frac{1}{\epsilon(x,y)}]^{-1}=\left[
\begin{array}{cc} 
\alpha(x_1) & \alpha(x_2) & ...& \alpha(x_p)& ...& \alpha(x_M)
\end{array}
\right]
$

```julia
#using ToeplitzMatrices, Test, LinearAlgebra
using ToeplitzMatrices, LinearAlgebra
using FFTW

# サンプリング数が偶数の時

P = 1

nn = Int(P * 2 + 1)

nnn = nn * nn

Af_a = zeros(Complex{Float64}, nnn, Rey)

#　関数作るときはサンプリング数の偶奇で場合分け

for i in 1 : Rey
    
    ## do fft xに沿っって
    Af = (1 / Rey) * fftshift(fft(F_r[:, i]'))

    Af_r = reverse(Af)

    #print(size(Af))
    
    o_0 = Int((length(Af) + 2) / 2)

    # o_0 is 0 order amplitude
    # N is 2 times of max order for Fourier 
    # P is max order for Fourier 

    N = Int(2 * P)

    # make the Toeplitz matrix center part is o_0
    
    Af_matrix = Toeplitz(Af_r[o_0 - 1 : o_0 - 1 + N], Af[o_0 : o_0 + N])

    #print(size(Af_matrix))

    II = Matrix{Complex{Float64}}(I, nn, nn)

    # calculate inverse of Toeplitz matrix
    
    Af_matrix = II / Af_matrix

    # 逆行列を 1行　 に並べる  nn * nn 要素数

    Af_a[:, i] = vec(Af_matrix)
    
end

#==
PyPlot.imshow(abs.(Af_a), cmap="viridis", origin="lower")
PyPlot.colorbar(label="Toeplitz")
PyPlot.xlabel("x")
PyPlot.ylabel("y")
PyPlot.grid(true, linestyle="dotted", color="white")
PyPlot.title("Toeplitz")
PyPlot.show()
==#
```

# (2N + 1)^2 by M
$ [\frac{1}{\epsilon(x,y)}]^{-1}=\left[
\begin{array}{cc} 
\alpha(y_1) \\ \alpha(y_2) \\ ...\\ \alpha(y_p)\\ ...\\ \alpha(y_M)
\end{array}
\right]=\left[
\begin{array}{cc} 
\alpha_{1,1}(y_1) & \alpha_{1,1}(y_2) & ..... & \alpha_{1,1}(y_{p}) & .....  & \alpha_{1,1}(y_{M-1}) & \alpha_{1,1}(y_M)\\ \alpha_{1,2}(y_1) & \alpha_{1,2}(y_2) & ..... & \alpha_{1,2}(y_{p}) & .....  &\alpha_{1,2}(y_{M-1}) & \alpha_{1,2}(y_M) \\ ...\\ \alpha_{1,2N+1}(y_1) & \alpha_{1,2N+1}(y_2) & ..... & \alpha_{1,2N+1}(y_{p}) & ..... &\alpha_{1,2N+1}(y_{M-1}) & \alpha_{1,2N+1}(y_M)\\ ...\\ \alpha_{2N+1,2N+1}(y_1) & \alpha_{2N+1,2N+1}(y_2) & .....& \alpha_{2N+1,2N+1}(y_{p}) & ..... & \alpha_{2N+1,2N+1}(y_{M-1}) & \alpha_{2N+1,2N+1}(y_M) 
\end{array}
\right]
$

# を計算したので次は上の行列の横方向(x方向に)フーリエ変換をし、2Nまでの字数を残す

```julia
# Af_a[:, i]でi列目を取り出す

size_A = size(Af_a)

#A_aa = fftshift(fft(Af_a, 1)) / Rey

#A_aa = fftshift(fft(Af_a, 1)) / Rey

A_aa = zeros(Complex{Float64}, nnn, Rex)

for i in 1 : nnn

    A_aa[i, :] = (1 / Rex) * fftshift(fft(Af_a[i, :]))
    
end

o_0 = Int((2 + size(A_aa)[2]) / 2)

N = Int(2 * P)

A_aa = A_aa[: , o_0　 - N  : o_0　 + N];

PyPlot.imshow(abs.(A_aa), cmap="viridis", origin="lower")
```

### (2N + 1)^2 by (4N+1)

A_aa = $\left[
\begin{array}{cc} 
\gamma_{1,1}^{-2N} & \gamma_{1,1}^{-2N+1} & ..... & \gamma_{1,1}^{-1} & \gamma_{1,1}^{0} & \gamma_{1,1}^{1} &..... & \gamma_{1,1}^{2N-1} & \gamma_{1,1}^{2N}\\ 
\gamma_{1,2}^{-2N} & \gamma_{1,2}^{-2N+1} & ..... & \gamma_{1,2}^{-1} & \gamma_{1,2}^{0} & \gamma_{1,2}^{1}&..... & 
\gamma_{1,2}^{2N-1} & \gamma_{1,2}^{2N}\\ 
...\\ 
\gamma_{2N+1,2N+1}^{-2N} & \gamma_{2N+1,2N+1}^{-2N+1} & .....& \gamma_{2N+1,2N+1}^{-1} & \gamma_{2N+1,2N+1}^{0} & \gamma_{2N+1,2N+1}^{1}&..... & \gamma_{2N+1,2N}^{2N-1} & \gamma_{2N+1,2N+1}^{2N}
\end{array}
\right]
$

### 横一列から取ってきて、Toeplitz Matrixを作成

```julia
# Toeplitz　行列作成 gamma_mnを作っていく
nn_1 = Int(4*P + 1)

#center
c_0 = Int((nn_1 + 1) / 2)

for i in 1:nnn

    # i 行目の要素等を取得
    gamma = A_aa[i, c_0:nn_1]
    
    gamma_re = reverse(A_aa[i, 1:c_0])
    
    gamma_i = Toeplitz(gamma_re, gamma)
    
    eval(Meta.parse("gamma_$i = $gamma_i"))
end

#nnn
```

### iは(2N+1)^2まで存在する (1,1),(1,2),...(2,1)...(2N+1,2N+1)の(2N+1)^2個

# 2N+1 by 2N+1

$\gamma_{m,n} = \left[
\begin{array}{cc} 
\gamma_{m,n}^{0} & \gamma_{m,n}^{-1} & ..... & \gamma_{m,n}^{-2N+1} & \gamma_{m,n}^{-2N}\\ 
\gamma_{m,n}^{1} & \gamma_{m,n}^{0} & ..... & \gamma_{m,n}^{-2N+2} & \gamma_{m,n}^{-2N+1} \\ 
...\\ 
\gamma_{m,n}^{2N} & \gamma_{m,n}^{2N-1} & ..... & \gamma_{n,m}^{1} & \gamma_{n,m}^{0}
\end{array}
\right]
$

# Block Matrix (2N +1)^2 by (2N +1)^2
$\left[
\begin{array}{cc} 
\gamma_{1,1} & \gamma_{1,2} & ..... &  ..... & \gamma_{1,2N+1}\\
\gamma_{2,1} & \gamma_{2,2} & ..... & \gamma_{2,2N} & \gamma_{2,2N+1} \\ 
...\\ 
\gamma_{2N+1,1} & \gamma_{2N+1,2} & ..... & ..... & \gamma_{2N+1,2N+1}
\end{array}
\right]
$

```julia
gamma_values = zeros(Complex{Float64}, nnn, nn, nn)

for i in 1:nnn

    # i 行目の要素等を取得
    gamma = A_aa[i, c_0:nn_1]
    
    gamma_re = reverse(A_aa[i, 1:c_0])
    
    gamma_i = Toeplitz(gamma_re, gamma)
    
    eval(Meta.parse("gamma_$i = $gamma_i"))

    gamma_values[i, :, :] = gamma_i 
end

gamma_matrix = [gamma_values[i, :, :] for i in 1:nnn];
```

```julia
function gamma_block_matrix(N::Int, matrices::Vararg{AbstractMatrix{T}}) where T
    
    if length(matrices) % N != 0
        throw(ArgumentError("The number of matrices must be a multiple of N"))
    end

    num_blocks = length(matrices) ÷ N  # ブロックの数を計算
    block_matrix = Matrix{T}(undef, size(matrices[1], 1) * num_blocks, size(matrices[1], 2) * N)

    for i in 1:num_blocks
        block_start = (i - 1) * N + 1
        block_end = i * N
        block = hcat(matrices[block_start:block_end]...)
        block_matrix[(i - 1) * size(block, 1) + 1:i * size(block, 1), :] = block
    end

    return block_matrix
end

create_block_matrix(matrices::Vector{Matrix{ComplexF64}}) = create_block_matrix(matrices...)
```

```julia
matrices = [gamma_matrix[i] for i in 1:length(gamma_matrix)]

gamma_block = gamma_block_matrix(nn, matrices...)

PyPlot.imshow(abs.(gamma_block), cmap="viridis", origin="lower")
PyPlot.colorbar(label="Toeplitz")
PyPlot.xlabel("x")
PyPlot.ylabel("y")
PyPlot.grid(true, linestyle="dotted", color="white")
PyPlot.title("Toeplitz")
PyPlot.show()
```

```julia
import LinearAlgebra

function create_block_matrix(matrices::Vararg{AbstractMatrix{T}}) where T
    # Get the dimensions of the first input matrix
    if isempty(matrices)
        throw(ArgumentError("At least one input matrix is required."))
    end
    matrix_size = size(matrices[1])

    num_matrices = length(matrices)

    # Check that all input matrices have the same dimensions
    for matrix in matrices
        if size(matrix) != matrix_size
            throw(ArgumentError("Input matrices must have the same shape."))
        end
    end

    # Calculate the size of the block matrix
    block_matrix = zeros(T, matrix_size[1] * num_matrices, matrix_size[2] * num_matrices)

    # Construct the block matrix
    for i in 1:num_matrices
        for j in 1:num_matrices
            block_matrix[(i-1)*matrix_size[1]+1:i*matrix_size[1], (j-1)*matrix_size[2]+1:j*matrix_size[2]] = matrices[(i+j-2) % num_matrices + 1]
        end
    end

    return block_matrix
end
```
